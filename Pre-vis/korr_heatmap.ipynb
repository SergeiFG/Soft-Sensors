```python
def heatmap(df_1):
    num_of_columns = df_1.shape[1] 
    df_1 = df_1.drop(df_1.columns[[num_of_columns - 1]], axis = 1)
    corr_matrix = df_1.corr()
    styled_df = corr_matrix.style.background_gradient(cmap='coolwarm')
    return display(styled_df)
```
